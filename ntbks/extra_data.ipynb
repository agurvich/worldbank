```python
foo = df[['Country Code','Monetary (%)', 'Educational attainment (%)',
       'Educational enrollment (%)', 'Electricity (%)', 'Sanitation (%)',
       'Drinking water (%)', 'Multidimensional poverty headcount ratio (%)',]].copy()


bar = df[['Country Code','FS_2015', 'FS_2016',
       'FS_2017', 'FS_2018', 'FS_2019', 'FS_2020', 'FS_2021', 'FS_2022',
       'FS_2023']].copy()

foo[foo.isna()]=None

bar[bar.isna()]=None



def transpose_df_on_key_1(df, key, orient='index'):
    # Replace NaN with None to ensure compatibility with JSON
    df_cleaned = df.where(pd.notnull(df), None)
    result_dict = df_cleaned.set_index(key).to_dict(orient=orient)
    return result_dict

def transpose_df_on_key_2(df, key, orient='list'):
    # Replace NaN with None to ensure compatibility with JSON
    df_cleaned = df.where(pd.notnull(df), None)
    result_dict = df_cleaned.set_index(key).T.to_dict(orient=orient)
    return result_dict

foo_result = transpose_df_on_key_1(foo,'Country Code')
bar_result = transpose_df_on_key_2(bar,'Country Code')

```

```python
foo[foo.isna()]=None
foo
```

```python
fs_years = [int(year.split('_')[1]) for year in 
    ['FS_2015', 'FS_2016','FS_2017', 'FS_2018', 'FS_2019', 'FS_2020', 'FS_2021', 'FS_2022','FS_2023']
                ]
output_data = {
    'fs_years':fs_years,
    'fs_data':bar_result
}
    
```

```python
with open('../app/public/data/mdp.json','w') as handle:
    json.dump(foo_result,handle)

with open('../app/public/data/foodsecurity.json','w') as handle:
    json.dump(output_data,handle)
```
